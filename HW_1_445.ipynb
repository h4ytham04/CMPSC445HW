```python
df[df.isna().any(axis=1)]
```

A, B, C, and H Distribution of data looks even enough to fill NaNs with mean values

```python
medianA = np.nanmedian(df["Feature A"])
medianB = np.nanmedian(df["Feature B"])
medianC = np.nanmedian(df["Feature C"])
medianH = np.nanmedian(df["Feature H"])

df["Feature A"] = df["Feature A"].fillna(medianA)
df["Feature B"] = df["Feature B"].fillna(medianB)
df["Feature C"] = df["Feature C"].fillna(medianC)
df["Feature H"] = df["Feature H"].fillna(medianH)

df[df.isna().any(axis=1)] #should be empty since all Nans filled with median
```

Data Set 2:

Pearson Correlation between ceatures and the target variable

```python
df2 = pd.read_csv('/Hw_1_data2.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df2
```

```python
acid_corr = df2[["acidity", "quality"]].corr()
volacid_corr = df2[["volatile acidity", "quality"]].corr()
citricacid_corr = df2[["citric acid", "quality"]].corr()
ressugar_corr = df2[["residual sugar", "quality"]].corr()
chloride_corr = df2[["chlorides", "quality"]].corr()
tsd_corr = df2[["total sulfur dioxide", "quality"]].corr()
density_corr = df2[["density", "quality"]].corr()
ph_corr = df2[["pH", "quality"]].corr()
sulphates_corr = df2[["sulphates", "quality"]].corr()
alc_corr = df2[["alcohol", "quality"]].corr()

correlation_dict = {
    "acidity": acid_corr.loc["acidity", "quality"],
    "volatile acidity": volacid_corr.loc["volatile acidity", "quality"],
    "citric acid": citricacid_corr.loc["citric acid", "quality"],
    "residual sugar": ressugar_corr.loc["residual sugar", "quality"],
    "chlorides": chloride_corr.loc["chlorides", "quality"],
    "total sulfur dioxide": tsd_corr.loc["total sulfur dioxide", "quality"],
    "density": density_corr.loc["density", "quality"],
    "pH": ph_corr.loc["pH", "quality"],
    "sulphates": sulphates_corr.loc["sulphates", "quality"],
    "alcohol": alc_corr.loc["alcohol", "quality"]
}

correlation_dict
```

Sort to see highest correlation to target

```python
{k: v for k, v in sorted(correlation_dict.items(), key=lambda item: item[1])}
```

Strongest Correlation is alcohol

```python
df2_corr = df2.corr()
df2_corr
```

```python
print("Non Redundant Features")
for i in range(len(df2_corr.columns)):
  for j in range(1 + i, len(df2_corr.columns)):
    if abs(df2_corr.iloc[i,j]) >= .51:
      print(f"{df2_corr.index[i]}, {df2_corr.columns[j]} have correlation {df2_corr.iloc[i, j]}")
```
